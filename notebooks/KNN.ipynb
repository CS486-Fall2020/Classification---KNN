```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from KNN import KNN
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.utils.estimator_checks import check_estimator
```

```python
#Heres the iris dataset for testing KNN on
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)

x_min, x_max = train_X[:, 0].min() - .5, train_X[:, 0].max() + .5
y_min, y_max = train_X[:, 1].min() - .5, train_X[:, 1].max() + .5

plt.figure(2, figsize=(10, 10))
plt.clf()

# Plot the training points
plt.scatter(train_X[:, 0], train_X[:, 1], c=train_y, cmap='jet',
            edgecolor='k', s=80)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

```python
scores = []
k = [x for x in range(1, 100)]

for x in range(1, 100):
    knn = KNN(x)
    knn.fit(train_X, train_y)
    predictions = knn.predict(test_X)[1]
    score = accuracy_score(test_y, predictions)
    scores.append(score)
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(k, scores)
plt.title("Performance of KNN on Iris dataset with respect to K", size=18)
plt.xlabel("K", size=15)
plt.ylabel("Accuracy score", size=15)
```
